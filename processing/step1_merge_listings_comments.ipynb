```python
import os
import numpy as np
import pandas as pd
from dataset. dataloader import Dataloader

os.getcwd()
os.chdir('..')
```

```python

data = Dataloader()
```

```python
listings = data.getListings()

print(listings.shape)
listings.head()
```

```python
comments = data.getCommentsFromListing(listings)
display(comments.head())
print(comments.shape)

```

# Merge comments with listings

```python
grouped_comments = comments.groupby('listing_id')['comments'].agg(list)
```

```python
grouped_comments.iloc[0]
```

```python
# the "id" column is the listings df is the same as the "listing_id" column in the comments df
# change the "id" column to "listing_id" to match the comments dataframe
listings.rename(columns={'id':'listing_id'}, inplace=True)
```

```python
# display all columns
pd.set_option('display.max_columns', None)
```

```python
# merge the comments with the listings df

df = listings.merge(grouped_comments, on='listing_id', how='left')
df.sample(3)
```

```python
df.comments.iloc[0]
```

```python
# delete the grouped_comments and listings dataframes to free up memory
del grouped_comments
del listings
del comments
```
